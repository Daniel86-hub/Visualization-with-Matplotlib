Fix: Dodge the bars, meaning, show them side by side To do this, we have to play with
the width and the positioning of the bars as shown in the next cell.
1. We want to reduce the width of the bars so that we can be able to fit them side by side on
the x axis.
2. Then shift the position of one bar by the width to be able to display the two bars side by
side.
We generate an array of x labels using the np.arange(len()) method. This turns the categorical
values on the x varaible(sector column) into an array of numbers which we can use to position the
minibars on the x axis.
By default the width of bars is 0.8. We set the width of the 2 Y variables to a smaller size (0.3),
then again shift one of the Y variables by that width, leaving the other in place. This gives us the
result we want.


```python
plt.title('Loan Amount by Sector',fontsize=15)
plt.xlabel('Sector')
plt.ylabel('Loan Amunt')
width=0.3
x_indices=np.arange(len(sector))
plt.bar(x_indices,loan,label='Loan Amount',width=width)
plt.bar(x_indices+width,fund,label='Funded Amount',width=width)
plt.xticks(ticks=x_indices,labels=sector,rotation=75)
plt.legend()
plt.show()
```

```python
plt.title('Loan Amount vs Lender Count', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 15)
plt.ylabel('Lender Count', fontsize = 15)

# convert the sector variable to a numerical array to be used so that the different markers are just different shades of that colour.

colour = np.arange(len(sector))
plt.xticks(rotation = 75)

plt.scatter(loan,lender,c=colour,marker='o',edgecolor='k',alpha=0.75)
plt.show()
```

You can map the colour output to one colour, so that the different markers are just different shades
of that colour.
We want to change the intesity of the color based on the sector

```python
plt.title('Loan Amount vs Lender Count', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 15)
plt.ylabel('Lender Count', fontsize = 15)

# convert the sector variable to a numerical array to be used so that the different markers are just different shades of that colour.
colour = np.arange(len(sector))
plt.xticks(rotation = 75)
plt.scatter(loan, lender, c = colour, cmap = 'Blues', marker = 'o', s = 100, edgecolor = 'k', alpha = 0.75)
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.title('Loan Amount vs Lender Count', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 15)
plt.ylabel('Lender Count', fontsize = 15)
colour = np.arange(len(sector))
plt.xticks(rotation = 75)
plt.scatter(loan, lender, c = colour, cmap = 'Blues', marker = 'o', s = loan,edgecolor = 'k', alpha = 0.75)
plt.show()
```

The loan amounts are too big, so we scale the values down by a factor of 1,000 for the markers to
fit well

```python
plt.figure(figsize = (10,5))
plt.title('Loan Amount vs Lender Count', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 15)
plt.ylabel('Lender Count', fontsize = 15)
colour = np.arange(len(sector))
plt.xticks(rotation = 75)
plt.scatter(loan, lender, c = colour, cmap = 'Blues', marker = 'o', s = loan/1000, edgecolor = 'k', alpha = 0.75)
plt.show()
```
